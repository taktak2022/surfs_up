# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
# MODULE 9.1.5 PREPARES THE DATABASE FILE TO BE CONNECTED TO LATER ON
# FEATURE OF SQLITE; SETS UP THE ABILITY TO QUERY A SQLITE DATABASE
engine = create_engine("sqlite:///hawaii.sqlite")
```

```python
# CREATE A BASE CLASS (FOUNDATION) FOR AN AUTOMAP SCHEMA
# reflect an existing database into a new model
Base = automap_base()

# reflect the SCHEMA from the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# MODULE 9.1.5 Create our session (link) from Python to the DB
session = Session(engine)
```

# Exploratory Climate Analysis

```python
# MODULE 9.2.1 
# Design a query to retrieve the last 12 months of precipitation data and plot the results. 
#Starting from the last data point in the database. 
prev_year = dt.date(2017, 8, 23) - dt.timedelta(days=365)
```

```python
# Calculate the date one year from the last date in data set.
# CREATE A VARIABLE TO STORE THE 'RESULTS' OF THE QUERY
results = []
```

```python
# Perform a query to retrieve the data and precipitation scores
# QUERY REFERENCING 2 PARAMETERS: THE DATE AND PRECIPITATION MEASUREMENT
results = session.query(Measurement.date, Measurement.prcp).filter(Measurement.date >= prev_year).all()
print(results)
```

```python
# MODULE 9.2.2 SAVE QUERY TO A PYTHON PANDAS DATAFRAME; CREATE VARIABLE 'df'
# Save the query results as a Pandas DataFrame and set the index to the date column
df = pd.DataFrame(results, columns=['date','precipitation'])
df
```

```python
# SET THE INDEX COLUMN BY DATE; REPLACE AUTO-GENERATED ROW NUMBERS WITH DATE COLUMN
df.set_index(df['date'], inplace=True)
print(df.to_string(index=False))
```

```python
# MODULE 9.2.3 
# Sort the dataframe by date
df = df.sort_index()
print(df.to_string(index=False))
```

```python
# MODULE 9.2.4; * SEE MODULE 5.1.5 FOR REFERENCE TO MATPLOTLIB BAR CHART CREATION (& GOOGLE) *
# * XTICK ORIENTATION WAS HORIZONTAL FOR SOME REASON; CHANGED TO VERTICAL WITH ADDITIONAL CODE *
# Use Pandas Plotting with Matplotlib to plot the data
df.plot(figsize=(15,5), color="green")
plt.title("Precipitation on Oahu August 2016- August 2017")
plt.xticks(rotation=90)
plt.show()
```

```python
# MODULE 9.2.5 GENERATE A STATISTICAL SUMMARY OF THE DATA
# Use Pandas to calculate the summary statistics for the precipitation data
df.describe()
```

```python
# MODULE 9.3.1 
# How many stations are available in this dataset?
session.query(func.count(Station.station)).all()
```

```python
# MODULE 9.3.2 SQLAlchemy QUERY OF STATIONS AND ITS DATA
# What are the most active stations?
# List the stations and the counts in descending order.
session.query(Measurement.station, func.count(Measurement.station)).\
group_by(Measurement.station).order_by(func.count(Measurement.station).desc()).all()
```

```python
# MODULE 9.3.3
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded and average temperature from the most active station. (WHICH IS WHERE ON OAHU??  HELLOOOO??)
session.query(func.min(Measurement.tobs), func.max(Measurement.tobs), func.avg(Measurement.tobs)).\
filter(Measurement.station == 'USC00519281').all()
```

```python
# MODULE 9.3.4
# Choose the station with the highest number of temperature observations.
# Query the last 12 months of temperature observation data for this station
results = session.query(Measurement.tobs).\
filter(Measurement.station == 'USC00519281').\
filter(Measurement.date >= prev_year).all()
print(results)
```

```python
# MODULE 9.3.4 CONVERT THE TEMPERATURE OBSERVATION RESULTS TO A DATAFRAME
df = pd.DataFrame(results, columns=['tobs'])
print(df)
```

```python
# MODULE 9.3.4 PLOT THE TEMPERATURE OBSERVATION RESULTS IN A HISTOGRAM
df.plot.hist(bins=13)
plt.tight_layout()
```
